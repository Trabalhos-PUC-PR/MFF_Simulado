A [Escala de Richter](https://pt.wikipedia.org/wiki/Escala_de_Richter) é a representação de uma função logaritmica usada para medir a magnitude de terremotos. A magnitude de um terremoto é relacionada com a quantidade de energia que é liberada durante o fenômeno. Instrumentos chamados de sismógrafos detectam o movimento da crosta terrestre. O movimento mais sutil que pode ser detectado por um sismógrafo é uma onda com amplitude $A_0$. 

Considerando:  $A$ como a medida da amplitude da onda de um terremoto e $A_0$ como a amplitude da menor onda detectável. Teremos o valor de Richter $R$ dado por: 

$$R = \log (\frac{A}{A_0})$$

##Exercício 2
Uma bomba hidráulica emite um ruído medido em 50 decibéis. Nesta mesma residência existe uma máquina de lavar que emite ruídos na magnitude de 62 decibéis. Quantas vezes o ruído da máquina de lavar é superior ao ruído da bomba hidráulica?

Resolução: 
Não é possível encontrar a relação entre estes ruidos por que a escala não é linear e não sabemos qual o valor do $P_0$, mas podemos colocar tudo em relação ao $P_0$. Então

$$d_{bomba}=10\log(\frac{B}{P_0})$$

Substituindo temos: 
$$50=10\log(\frac{B}{P_0})$$
$$5=\log(\frac{B}{P_0})$$
$$10^5=\frac{B}{P_0}$$
$$B=10^5P_0$$

O mesmo pode ser feito com a náquina de lavar. 

$$d_{máquina}=10\log(\frac{M}{P_0})$$

Substituindo temos: 
$$62=10\log(\frac{M}{P_0})$$
$$6.2=\log(\frac{M}{P_0})$$
$$10^{6.2}=\frac{M}{P_0}$$
$$M=10^{6.2}P_0$$

Agora podemos saber a relação entre M e B

 


```python
M = 10**6.2
B = 10**5
print('A relação é: ', M/B)
#ou 10**1,2
```

Você também pode pensar este problema considerando a formula inteira... 

$$d_{máquina}=10\log(\frac{M}{P_0})$$
$$d_{bomba}=10\log(\frac{B}{P_0})$$

removendo o $10$ temos: 

$$\frac{d_{máquina}}{10}=\log(\frac{M}{P_0})$$
$$\frac{d_{bomba}}{10}=\log(\frac{B}{P_0})$$

Removendo os log

$$10^{\frac{d_{máquina}}{10}}=(\frac{M}{P_0})$$
$$10^{\frac{d_{bomba}}{10}}=(\frac{B}{P_0})$$

E a gente quer a relação entre estas duas equações então: 

$$\frac{10^{\frac{d_{máquina}}{10}}}{10^{\frac{d_{bomba}}{10}}} = \frac{(\frac{M}{P_0})}{(\frac{B}{P_0})}$$

simplificando temos: 

$$10^{\frac{d_{máquina}-d_{bomba}}{10}} = \frac{M}{B} $$

Substituíndo temos: 


$$10^{\frac{62-50}{10}} = \frac{M}{B} $$

$$10^{1.2} = \frac{M}{B} $$


##Exercício 3
Um bom exemplo do uso de funções exponenciais é o cálculo de juros compostos. Levando este modelo em consideração. Determine quanto tempo será necessário, em anos, para que um determinado montante em dinheiro dobre se for investido a taxas de juros de 0,5% ao mês.




Resposta: 
Em uma de nossas aulas chegamos a fórmula para cálculo dos juros compostos na equação: 
<p>&nbsp;</p>
$$Montante = ValorInicial*(1+jurosEmDecimal)^{tempo}$$
<p>&nbsp;</p>

Vamos simplificar um pouco mais 
<p>&nbsp;</p>
$$M = V*(1+j)^{t}$$
<p>&nbsp;</p>
O que a gente precisa é encontrar o tempo para que $M = 2V$

Podemos colocar em forma de equação:
<p>&nbsp;</p>
$$2V = V*(1+j)^{t}$$
<p>&nbsp;</p>
$$0 = V*(1+j)^{t} - 2V$$


```python
V = sp.Symbol('V')
t = sp.Symbol('t')
#Podemos usar o solve para não nos preocuparmos com a algebra: em meses teremos
sp.solve(V*(1+(0.5/100))**t-2*V,t)
#mas, observe que o resulta do solv é uma lista logo 
#se pegarmos este resultado no índice zero e dividirmo por doze teresmo em anos: 
print(sp.solve(V*(1+(0.5/100))**t-2*V,t)[0]/12)
```

##Exercício 4
A População mundial é de 7,3 bilhões de pessoas e cresce a uma taxa positiva de 85 milhões de pessoas a cada ano. Considerando estes dados, crie um modelo computacional capaz de indicar a população mundial no ano de 2050. Uma vez que este modelo tenha sido determinado, indique quantos anos serão necessários, a partir de 2021, para que a população mundial dobre.

```python
#Este é o memos modelo que já vimos várias vezes... crescimento exponencial. 
#Podemos começar vendo a relação da taxa de crescimento com em relação a população. 
#Isso pode ser encontrado com uma regra de 3
taxa = (85e6)/7.3e9
print(taxa)
#como o crescimento é positivo teremos 
crescimento = 1+taxa
print(crescimento)
```

Assim, nossa população em um determinado período será

$P=P{inicial}*crecimento^tempo$

Para encontrar a população no ano de 2050 teremos $2050-2021$ logo

```python
P = 7300000000*crescimento**(2050-2021)
print(P)
```

Para ver quanto tempo levaremos para dobrar a população mundial podemos: 

$$7300000000*2 = 7300000000*crescimento^t$$

Ou

$$2 = crescimento^t$$

$$\log 2 = \log crescimento^t$$

$$\log 2 = t\times \log crescimento$$

$$t = \frac{\log 2}{\log crescimento}$$


$$$$

```python
t = math.log(2)/math.log(crescimento)
print(t)
```
