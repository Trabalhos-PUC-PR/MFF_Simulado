A [Escala de Richter](https://pt.wikipedia.org/wiki/Escala_de_Richter) é a representação de uma função logaritmica usada para medir a magnitude de terremotos. A magnitude de um terremoto é relacionada com a quantidade de energia que é liberada durante o fenômento. Instrumentos chamados de sismógrafos detectam o movimento da crosta terrestre. O movimento mais sutil que pode ser detectado por um sismógrafo é uma onda com amplitude $A_0$. 

Considerando:  $A$ como a medida da amplitude da onda de um terremoto e $A_0$ como a amplitude da menor onda detectável. Teremos o valor de Richter $R$ dado por: 

$$R = \log (\frac{A}{A_0})$$

##Exercício 1
Um terremoto foi medido com tendo uma amplitude de onda 392 vezes maior que $A_0$. Neste caso, qual seria a magnitude deste evento na Escala de Richter? Caso a amplitude fosse de 3920, qual seria esta magnitude? Qual a magnitude de um terremoto cuja amplitude medida seja $A_0$? Trace um gráfico que permita relacionar a amplitude da onda medida e o Fator de Richter, lembrando que este gráfico é a representação da Escala de Richter e que ele deve começar em amplitudes iguais a $A_0$ 


```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt 

A0 = sp.Symbol('A0')

def escala(A):
    return sp.log(A/A0, 10) #log de (A/A0) na base 10

A1 = 392 * A0
R1 = escala(A1)

A2 = 3920
R2 = escala(A2)

A3 = A0
R3 = escala(A3)

sp.plot(R1, line_color='red', xlim=(-10,10), ylim=(-10,10))
sp.plot(R2, line_color='blue', xlim=(-5,5), ylim=(-10,10))
sp.plot(R3, line_color='green', xlim=(-10,10), ylim=(-10,10))
```

O som é medido em uma escala logarítmica chamada de [Decibel](https://pt.wikipedia.org/wiki/Decibel) cuja equação apresenta uma forma próxima a da equação da Escala de Richter: 

$$d=10\log(\frac{P}{P_0})$$

Onde $P$ é a intensidade, ou magnitude, de uma onda sonora e $P_0$ é a menor intensidade de som que pode ser percebida pelo ouvido humano. 

##Exercício 2
Uma bomba hidráulica emite um ruído medido em 50 decibéis. Nesta mesma residência existe uma maquina de lavar que emite ruídos na magnitude de 62 decibéis. Quantas vezes o ruído da máquina de lavar é superior ao ruído da bomba hidráulica?

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt 

P0 = sp.Symbol('P0')

def decibeis(P):
    return 10 * sp.log(P/P0, 10)

P1 = 50
D1 = decibeis(P1)

P2 = 62
D2 = decibeis(P2)

print(D2 / D1)

sp.plot(D2/D1, line_color='red', xlim=(-10,10), ylim=(-2,2))
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt 

T = 0
J = 0.005
C = 1000
Montante = 0


while(Montante <= C*2):
    T += 1
    Montante = C * (1 + J) ** T

print(T,"Meses, com o valor de aproximadamente ",Montante,"cascalhitos")
#sp.plot(juros(C), line_color='red', xlim=(-15,15), ylim=(0, C*2))
```
